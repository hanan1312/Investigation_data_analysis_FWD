## Dataset Overview

> I took my data from this website "https://divvy-tripdata.s3.amazonaws.com/index.html" after cleaning The data contains 1698725 records and 16 columns, for the bike sharing system in year 2019, the most numerical data that matters are the (tripduration, and age) .. with categorical data in (gender, and usertype) and string data in the rest..I stored my new data after cleaning in a csv file called (new_data.csv) to work from it directly here.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("data/new_data.csv")
```

## Relation between days of the weeks and the usertype 

> It seems that subscribers rides more the whole week in work days and less in weekends, on the other hand customers rides more in weekends

```python
sb.countplot(data=df, x='start_day', hue='usertype');
plt.xlabel('Day of The Week', fontsize=20);
plt.xticks(rotation=45);
plt.ylabel('Count', fontsize=20);
```

## Relation between age and months of the year 

> seems the riders with young age bike in summer than elders that rides in all months including winter

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=[10, 8])
sb.boxplot(data=df, x='start_month', y='age', color=base_color);
plt.xlabel('Month of The Year', fontsize=20);
plt.ylabel('Age', fontsize=20);
```

## Relation between age and gender according to days of the week

> Females as well as males like riding in work days more than week end .. but male's population is more than female's in using the system, on the other hand females are mostly younger than males.

```python
sb.countplot(data = df, x='start_day', hue='gender')
plt.xticks(rotation=45);
plt.xlabel('Relation between gender & the day of the week', fontsize=20);
```

## Relation between age and gender

> females in young age is more than the age of males using the system.customers are younger in age than subscribers..most of the riders are in the range of 20 - 40 in age mostly distributed in the whole week except the weekend but, thoese who rides in weekends are mostly the younger ones...Here the higher durationtrips are for males more than females, but the females are mostly younger than males. For different user types both are showing similer trends for age and tripduration. but it seems that customers rides durations more than subscribers.

```python
plt.figure(figsize=[10, 8])
base_color = sb.color_palette()[0]
sb.boxplot(data=df, x='gender', y='age', color=base_color);
plt.xlabel('gender', fontsize=20);
plt.ylabel('Age', fontsize=20);
```

```python
plt.figure(figsize=[10, 8])
sb.boxplot(data=df, x='usertype', y='age', color=base_color);
plt.xlabel('usertype', fontsize=20);
plt.ylabel('Age', fontsize=20);
```

```python
plt.figure(figsize=[10, 8])
sb.boxplot(data=df, x='start_day', y='age', color=base_color);
plt.xlabel('Day of The Week', fontsize=20);
plt.ylabel('Age', fontsize=20);
```

```python
user_type_markers = [['Customer', 's'],['Subscriber', 'o']]

for utype, marker in user_type_markers:
    df_utype = df[df['usertype'] == utype]
    plt.scatter((df_utype['age']), df_utype['tripduration'], marker = marker, alpha=0.25)
plt.legend(['Customer','Subscriber'])
plt.axis([10, 80, -500, 9000 ])
plt.xlabel('Age (year)')
plt.ylabel('Duration (sec)')
plt.show()
```

## A summary

> As we can see a lot more subscribers using the bike sharing system than casual customers, both of them prefer ride the most during the summer season and the least during the winter months. Obviously different there are usage patterns and riding habits are observed between the two types of riders. Subscribers used the system heavily on work days, but according to gender males are using the system more than females, females mostly young although males older, both of genders prefere riding in work days than weekends, but after all old riders use the system in all seasons on the other hand younger ones prefer summer season more.
